```python
genderData = ["LAKI-LAKI", "PEREMPUAN"]
jumlahData=[jumlah_kL, jumlah_kP]
warna=["#1f77b4","#d62728"]


plt.title("Perbandingan jumlah kasus")
plt.pie(jumlahData, labels=genderData, colors=warna, startangle=90, shadow=False, explode= (0, 0.1),radius=1.2, autopct='%1.1f%%')
plt.legend()
plt.show
```
