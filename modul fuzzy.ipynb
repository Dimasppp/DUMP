2.	Buatlah program untuk mengubah nilai crips suhu dan jumlah orang sesuai dengan fungsi keanggotaan yang sudah Anda buat pada nomor 1 dengan masukan variabel suhu 26°C dan jumlah orang 25 orang.

Suhu

```python
def suhuTinggi(suhu):
    if suhu > 26:
        return 1
    elif suhu < 26:
        return 0
    else:
        return ((suhu-26)/(28-27))
# suhuTinggi(27)
```

```python
def suhuNormal(suhu):
    if suhu >= 26 and suhu <= 27:
        return (27 - suhu)/(28-27)
    elif suhu == 27:
        return 1
    elif suhu >= 27 and suhu <= 28:
        return(suhu - 27)/ (28 - 27)
    else :
        return 0
```

```python
def suhuRendah(suhu):
    if suhu < 27:
        return 1
    elif suhu > 27:
        return 0
    else:
        return ((27-suhu)/(28-27))
# suhuRendah(27)
```

Jumlah

```python
def jumlahSedikit(jumlah):
    if jumlah <= 17:
        return 1
    elif jumlah <= 26 and jumlah >= 25:
        return ((26-jumlah)(26-25))
    else:
        return 0
```

```python
def jumlahIdeal(jumlah):
    if jumlah >= 14 and jumlah <= 22:
        return ((15-jumlah)/(15-14))
    elif jumlah >= 15 and jumlah <= 22:
        return 1
    elif jumlah >= 21 and jumlah <= 22:
        return ((jumlah-21)/(23-22))
    else:
        return 0
```

```python
def jumlahBanyak(jumlah):
    if jumlah >= 26:
        return 1
    elif jumlah <= 26 and jumlah >= 25:
        return ((26-jumlah)/(26-25))
    else :
        return 0
```

3.	Jika aturan-aturan yang digunakan dalam sistem tersebut adalah:
a.	Jika suhu tinggi dan jumlah orang sedikit maka AC dibiarkan
b.	Jika suhu tinggi dan jumlah orang ideal maka hidupkan AC
c.	Jika suhu tinggi dan jumlah orang banyak maka hidupkan AC
d.	Jika suhu normal dan jumlah orang sedikit maka AC dibiarkan
e.	Jika suhu normal dan jumlah orang ideal maka AC dibiarkan
f.	Jika suhu normal dan jumlah orang banyak maka hidupkan AC
g.	Jika suhu rendah dan jumlah orang sedikit maka matikan AC
h.	Jika suhu rendah dan jumlah orang ideal maka matikan AC
i.	Jika suhu rendah dan jumlah orang banyak maka AC dibiarkan
Aturan apa saja yang aktif berdasarkan masukan seperti jawaban nomor 2.


Alpha Predicate

```python
alpha_predicate = []
suhu = float(input("Suhu : "))
jumlah = int(input("Jumlah orang : "))

a = min(suhuTinggi(suhu),jumlahSedikit(jumlah))
b = min(suhuTinggi(suhu),jumlahIdeal(jumlah))
c = min(suhuTinggi(suhu),jumlahBanyak(jumlah))
d = min(suhuNormal(suhu),jumlahSedikit(jumlah))
e = min(suhuNormal(suhu),jumlahIdeal(jumlah))
f = min(suhuNormal(suhu),jumlahBanyak(jumlah))
g = min(suhuRendah(suhu),jumlahSedikit(jumlah))
h = min(suhuRendah(suhu),jumlahIdeal(jumlah))
i = min(suhuRendah(suhu),jumlahBanyak(jumlah))

alpha_predicate.append(a)
alpha_predicate.append(b)
alpha_predicate.append(c)
alpha_predicate.append(d)
alpha_predicate.append(e)
alpha_predicate.append(f)
alpha_predicate.append(g)
alpha_predicate.append(h)
alpha_predicate.append(i)
print(f"Alpha predicate adalah {alpha_predicate}")
```

4.	Buatlah program untuk melakukan proses inferensi Tsukamoto untuk penggambilan keputusan berdasarkan aturan yang aktif di nomor 3 dan masukan dari nomor 2

```python
def Biarkan(x):
    z = (2-(x*(2-1)))
    z1 = (x*(x-8)+8)
    return max(z,z1)

def Dimatikan(x):
    return (2-(x*(2-1)))

def Dinyalakan(x):
    return (x*(x-8)+8)
```

```python
zValue = []
zValue.append(Biarkan(a))
zValue.append(Dinyalakan(b))
zValue.append(Dinyalakan(c))
zValue.append(Biarkan(d))
zValue.append(Biarkan(e))
zValue.append(Dinyalakan(f))
zValue.append(Dimatikan(g))
zValue.append(Dimatikan(h))
zValue.append(Biarkan(i))
print (zValue)
```
